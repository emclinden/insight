```python
dbname = 'coffee_detai1'
username = 'emclinden'
engine = create_engine('postgres://%s@localhost/%s'%(username,dbname))
print engine.url
```

```python
query = "SELECT * from image_stats_del" #% patient
qr=pd.read_sql_query(query,engine)
```

```python
x1 = qr.brightness_scale; x2 = qr.contrast_scale; x3 = qr.ntags_scale; x4 = qr.comments_scale;
x =np.column_stack((x1,x2,x3,x4))
y = qr.result_y
```

```python
nfavs = qr.n_favs.astype('str').astype('int').astype('float')
nviews = qr.n_views.astype('str').astype('int').astype('float')
yy = nfavs / nviews
```

```python
xlearn = x[0:3000,:]
ylearn = y[0:3000]
xval = x[3000:,:]
yval = y[3000:]
```

```python
clf = LinearRegression()
clf.fit (xlearn, ylearn)
```

```python
print clf.coef_
print clf.intercept_
```

```python
ypred = clf.predict(xval)
```

```python
newdf = pd.DataFrame()
newdf['picture_id'] = qr.picture_id[3000:]
newdf['bright'] = qr.brightness_scale[3000:]
newdf['contrast'] = qr.contrast_scale[3000:]
newdf['ntags'] = qr.ntags_scale[3000:]
newdf['comments'] = qr.comments_scale[3000:]
newdf['ypredict'] = ypred
newdf['yactual'] = yval
newdf.to_sql('valset_865', engine, if_exists='replace')
```
